```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
df
```

```python
df.info()
```

```python
df["Cabin"].unique()
```

```python
df["Cabin"].isna().sum()
```

```python
df["Cabin"] = df["Cabin"].fillna("Not")
```

```python
df["Embarked"].value_counts()
```

```python
df["Embarked"] = df["Embarked"].fillna("S")
```

```python
df["Sex"] = df["Sex"].apply(lambda x : 1 if x == "male" else 0)
```

```python
df["Ticket"].unique()
```

```python
df["Ticket"] = df["Ticket"].apply(lambda x: x if len(x.split())==1 else x.split()[1])
```

```python
df = df.drop(columns=["new"], axis= 1)
```
